```
#imports
import numpy
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from PIL import Image, ImageFilter
from numpy import array

img1 = Image.open("/content/drive/My Drive/Imagens_Mario/bailey-zindel-NRQV-hBF10M-unsplash.jpg")
img2 = Image.open("/content/drive/My Drive/Imagens_Mario/claudio-testa--SO3JtE3gZo-unsplash.jpg")
img3 = Image.open("/content/drive/My Drive/Imagens_Mario/johannes-plenio-RwHv7LgeC7s-unsplash.jpg")
```

# 1a parte

No espaço de código abaixo, plote as 3 imagens, uma ao lado da outra. Ainda, plote o histograma de cada canal de cor (R - Vermelho, G - Verde e B - Azul) abaixo da respectiva imagem.

```
img1_arr, img2_arr, img3_arr = np.array(img1), np.array(img2), np.array(img3)

red1, red2, red3 = img1_arr[:,:,0], img2_arr[:,:,0], img3_arr[:,:,0]
green1, green2, green3 = img1_arr[:,:,1], img2_arr[:,:,1], img3_arr[:,:,1]
blue1, blue2, blue3  = img1_arr[:,:,2], img2_arr[:,:,2], img3_arr[:,:,2]

def plotar_imagem(p1, p2, p3, imagem):
  plt.subplot(p1, p2, p3)
  plt.axis('off')
  plt.imshow(imagem)

def plotar_histograma(p1, p2, p3, p4, p5, red, green, blue):
  plt.subplot(p1, p2, p3) 
  plt.hist(red.ravel(), color='red')
  plt.subplot(p1, p2, p4) 
  plt.hist(green.ravel(), color='green')
  plt.subplot(p1, p2, p5) 
  plt.hist(blue.ravel(), color='blue')

plt.figure(figsize=(20,30))

plotar_imagem(6, 3, 1, img1)
plotar_histograma(6, 3, 4, 7, 10, red1, green1, blue1)

plotar_imagem(6, 3, 2, img2)
plotar_histograma(6, 3, 5, 8, 11, red2, green2, blue2)

plotar_imagem(6, 3, 3, img3)
plotar_histograma(6, 3, 6, 9, 12, red3, green3, blue3)

plt.show()
```

Olhando para os gráficos gerados, discorram comparando os resultados.



> Os histogramas mostram valores mais altos quando as imagens possuem dominância de um determinado tipo de cor. Por exemplo, a imagem que possui tons mais azulados possui mais pontos com valores de 255 no histograma se comparado as imagens com menos tons de azul. O mesmo ocorre para os valores de vermelho e de verde que aparecem mais quando são dominantes em determinadas partes da imagem.

Vamos gerar algumas imagens preto e branco, utilizando limiares diferentes:

*  45
*  100
*  145
*  200
*  245

Escreva o código no espaço abaixo.

```
def preto_e_branco(limiar):
  imagem_pb = np.array(img1)
  img = imagem_pb
  img[img[:,:,0] > limiar] = 255
  img[img[:,:,0] <= limiar] = 0
  img[img[:,:,1] > limiar] = 255
  img[img[:,:,1] <= limiar] = 0
  img[img[:,:,2] > limiar] = 255
  img[img[:,:,2] <= limiar] = 0
  return img

imagem_pb_45 = preto_e_branco(45)
imagem_pb_100 = preto_e_branco(100)
imagem_pb_145 = preto_e_branco(145)
imagem_pb_200 = preto_e_branco(200)
imagem_pb_245 = preto_e_branco(245)
```

Plote as imagens geradas. (Escreva o código no espaço abaixo)

```
plt.figure(figsize=(20,30))

plotar_imagem(6, 3, 1, imagem_pb_45)
plotar_imagem(6, 3, 2, imagem_pb_100)
plotar_imagem(6, 3, 3, imagem_pb_145)
plotar_imagem(6, 3, 4, imagem_pb_200)
plotar_imagem(6, 3, 5, imagem_pb_245)

plt.show()
```

> O valor do limiar estabelece quais valores irão assumir as cores pretas e brancas. Por exemplo, se o limiar for um valor alto e os canais RGB possírem muitos valores baixos, a imagem ficará com mais área preta do que branca.
